```python
from aeon.datasets import load_classification
from tscglue.models import LokyStackerV8Base
```

```python
X_train, y_train = load_classification("BasicMotions", split="train")
X_test, y_test   = load_classification("BasicMotions", split="test")

print(f"train: {X_train.shape}, test: {X_test.shape}")

model = LokyStackerV8Base(random_state=492, n_repetitions=1, n_jobs=16, keep_features=True, verbose=2)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

from sklearn.metrics import accuracy_score
print(f"accuracy: {accuracy_score(y_test, y_pred):.4f}")
```
